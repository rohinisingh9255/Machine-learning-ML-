# Q2) Extract data from outcome column is a variable named y

```python
y = data.iloc[:,-1]
y.head()
```

# Q3) Extract data from every column except outcome column in a variable name X

```python
X= data.iloc[:,0:8]
X.head()
```

# Q4) Divide the data set into two part training and testing in 70% and 30% proportion.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42,test_size=0.30)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Q5) Create the train Logistic Regression Model on Training set.

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

# Q6) Make prediction based on the testing set using the trained model.      Q7) Check the performance by calculating the confusion matrix and accuracy score of the model.

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
conf_matrics = confusion_matrix(y_test,y_pred)
```

```python
conf_matrics
```

```python
score = lr.score(X_test, y_test)
print(score)
```
